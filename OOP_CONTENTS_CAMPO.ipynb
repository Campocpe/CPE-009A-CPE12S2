```python
#Animal's Features like name, habitat, and temperature
class Animals:
  def __init__(self,name,habitat,temperature):
    self.name = name
    self.habitat = habitat
    self.temperature = temperature


  def features(self):
    return f"The {self.name} is a {self.temperature}-blooded animal that lives on {self.habitat}."

mammals = Animals("Dog", "Land", "Warm")
birds = Animals("Sparrow", "Air", "Warm")
reptiles = Animals("Snake", "Terrestrial", "Cold")

print(mammals.name)
print(mammals.habitat)

print(birds.name)
print(birds.habitat)

print(reptiles.name)
print(reptiles.habitat)



print(mammals.features())
print(birds.features())
print(reptiles.features())
```

```python
class Shape:

    def __init__(self, name, sides):
        self.name = name
        self.sides = sides

    def description(self):
        return f"The {self.name} has {self.sides} sides."


    def area_circle(self, radius):
        return 3.14 * (radius ** 2)

    def area_square(self, side):
        return side * side

    def area_rectangle(self, length, width):
        return length * width

    def area_triangle(self, base, height):
        return 0.5 * base * height


obj1 = Shape("Square", 4)
obj2 = Shape("Circle", 0)
obj3 = Shape("Rectangle", 4)
obj4 = Shape("Triangle", 3)

print(obj1.description())
print(f"Area of {obj1.name}: {obj1.area_square(5)}")

print(obj2.description())
print(f"Area of {obj2.name}: {obj2.area_circle(10)}")

print(obj3.description())
print(f"Area of {obj3.name}: {obj3.area_rectangle(5, 10)}")

print(obj4.description())
print(f"Area of {obj4.name}: {obj4.area_triangle(10, 8)}")
```
