Develop a python scripts that:

Calculate total marks and percentage
Assign grade for each subject based on marks:
90-100: A+

80-89: A

70-79: B+

60-69: B

50-59: C

40-49: D

Below 40: F (Fail)

Determine overall result (Pass/Fail)


Performance Remarks:


Percentage ≥ 90: "Outstanding"

Percentage ≥ 80: "Excellent"

Percentage ≥ 70: "Very Good"

Percentage ≥ 60: "Good"

Percentage ≥ 40: "Satisfactory"

Below 40: "Needs Improvement"

```python
#course 1
exam = float
quiz = float
hoa = float
course = (input("enter course:"))
print ("calculating the grades for " + course)

hoatotal = float(input("enter hoa total percentage(%): "))
quiztotal = float(input("enter quiz total percentage(%): "))
examtotal = float(input("enter exam total percentage(%): "))
grade = float ((examtotal + quiztotal + hoatotal) /3)
print (course + " grade is ")
print (grade, "%")


if grade >= 90:
  print ("A+")
elif grade >= 80:
  print ("A")
elif grade >= 70:
  print ("B+")
elif grade >= 60:
  print ("B")
elif grade >= 50:
  print ("C")
elif grade >= 40:
  print ("D")
else:
  print ("F")







if grade >= 90:
    print ("Outstanding")
elif grade >= 80:
    print ("Excellent")
elif grade >= 70:
    print ("Very Good")
elif grade >= 60:
    print ("Good")
elif grade >= 40:
    print ("Satisfactory")
else:
    print ("Needs Improvement")

if grade < 40:
    print ("Fail")
else:
    print ("Pass")

#course 2
exam = float
quiz = float
hoa = float
course = (input("enter course:"))
print ("calculating the grades for " + course)

hoatotal = float(input("enter hoa total percentage(%): "))
quiztotal = float(input("enter quiz total percentage(%): "))
examtotal = float(input("enter exam total percentage(%): "))
grade = float ((examtotal + quiztotal + hoatotal) /3)
print (course + " grade is ")
print (grade, "%")


if grade >= 90:
  print ("A+")
elif grade >= 80:
  print ("A")
elif grade >= 70:
  print ("B+")
elif grade >= 60:
  print ("B")
elif grade >= 50:
  print ("C")
elif grade >= 40:
  print ("D")
else:
  print ("F")







if grade >= 90:
    print ("Outstanding")
elif grade >= 80:
    print ("Excellent")
elif grade >= 70:
    print ("Very Good")
elif grade >= 60:
    print ("Good")
elif grade >= 40:
    print ("Satisfactory")
else:
    print ("Needs Improvement")

if grade < 40:
    print ("Fail")
else:
    print ("Pass")


#course 3
exam = float
quiz = float
hoa = float
course = (input("enter course:"))
print ("calculating the grades for " + course)

hoatotal = float(input("enter hoa total percentage(%): "))
quiztotal = float(input("enter quiz total percentage(%): "))
examtotal = float(input("enter exam total percentage(%): "))
grade = float ((examtotal + quiztotal + hoatotal) /3)
print (course + " grade is ")
print (grade, "%")


if grade >= 90:
  print ("A+")
elif grade >= 80:
  print ("A")
elif grade >= 70:
  print ("B+")
elif grade >= 60:
  print ("B")
elif grade >= 50:
  print ("C")
elif grade >= 40:
  print ("D")
else:
  print ("F")







if grade >= 90:
    print ("Outstanding")
elif grade >= 80:
    print ("Excellent")
elif grade >= 70:
    print ("Very Good")
elif grade >= 60:
    print ("Good")
elif grade >= 40:
    print ("Satisfactory")
else:
    print ("Needs Improvement")

if grade < 40:
    print ("Fail")
else:
    print ("Pass")

#course 4
exam = float
quiz = float
hoa = float
course = (input("enter course:"))
print ("calculating the grades for " + course)

hoatotal = float(input("enter hoa total percentage(%): "))
quiztotal = float(input("enter quiz total percentage(%): "))
examtotal = float(input("enter exam total percentage(%): "))
grade = float ((examtotal + quiztotal + hoatotal) /3)
print (course + " grade is ")
print (grade, "%")


if grade >= 90:
  print ("A+")
elif grade >= 80:
  print ("A")
elif grade >= 70:
  print ("B+")
elif grade >= 60:
  print ("B")
elif grade >= 50:
  print ("C")
elif grade >= 40:
  print ("D")
else:
  print ("F")







if grade >= 90:
    print ("Outstanding")
elif grade >= 80:
    print ("Excellent")
elif grade >= 70:
    print ("Very Good")
elif grade >= 60:
    print ("Good")
elif grade >= 40:
    print ("Satisfactory")
else:
    print ("Needs Improvement")

if grade < 40:
    print ("Fail")
else:
    print ("Pass")
```
