```python
#Nested loops
adj = ["red","big","tasty"]
fruits = ["apple","banana","cherry"]

for x in adj:
  for y in fruits:
    print(x,y)
```

```python
#The break statement

i = 1
while i < 6:
  print(i)
  if i == 3:
    break
  i += 1
```

```python
#The continue statement

i = 0
while i < 6:
  i += 1
  if i == 3:

    continue
  print(i)
```

```python
#The else statement

i = 1
while i < 6:
  print(i)
  i += 1
else:
  print("i is no longer less than 6")
```
