```python
#1
coordinates_list = [(1, 1), (2, 3)]

x1 = coordinates_list[0][0]
y1 = coordinates_list[0][1]
x2 = coordinates_list[1][0]
y2 = coordinates_list[1][1]

dx = x2 - x1
dy = y2 - y1

distance = (dx**2 + dy**2)**0.5

print(f"The distance between the points is: {distance}")
```

```python
#2
text = "That is a stupid mistake."
target = "stupid"

stars = "*" * len(target)

filtered_text = text.replace(target, stars)
print(filtered_text)
```

```python
#3
phonebook = [
    {"id": 1, "name": "jath", "contact": "555-0101", "address": "7 narcisus st. "}, {"id": 2, "name": "nicholas", "contact": "555-0202", "address": "1480 jalan st."}, {"id": 3, "name": "janerio", "contact": "555-0303", "address": "14c imdelda village "}
]


search_id = int(input("Enter the ID of the record: "))


for i in range(len(phonebook)):

    record = phonebook[i]


    if record["id"] == search_id:
        print(f"Name: {record['name']}")
        print(f"Contact: {record['contact']}")
        print(f"Address: {record['address']}")
        break
else:

    print("Record not found.")
```
