```python
#Nested IF
x = 41

if x > 10:
  print("Above 10,")
  if x > 20:
    print("and also above 20!")
    if x > 30:
      print("and also above 30!")
      if x > 40:
        print("and also above 40!")
        if x > 50:
          print("and also above 50!")
        else:
          print("but not above 50.")
      else:
        print("but not above 40.")
    else:
      print("but not above 30")
  else:
    print("but not above 20.")
else :
  print("not above 10")
```
